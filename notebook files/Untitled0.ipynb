```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
models = ["Logistic Regression", "Random Forest", "LDA", "Decision Tree", "Ensemble Model 1", "Ensemble Model 2"]
metrics = ["Accuracy", "Precision", "Recall", "F-1 Score"]
scores = np.array([
    [0.7643, 0.6786, 0.6905, 0.6548, 0.7976, 0.75],    # Accuracy
    [0.7425, 0.6995, 0.6901, 0.6538, 0.7993, 0.7894],  # Precision
    [0.7643, 0.6786, 0.6905, 0.6548, 0.7976, 0.75],    # Recall
    [0.745, 0.6671, 0.6882, 0.6497, 0.7709, 0.7169]    # F1-Score
])

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharex=True)
axes = axes.flatten()

# Plotting each metric as a separate subplot
for i, metric in enumerate(metrics):
    axes[i].bar(models, scores[i], color='skyblue', edgecolor='black')
    axes[i].set_title(metric)
    axes[i].set_ylabel("Score")
    axes[i].set_ylim(0.6, 0.85)
    axes[i].tick_params(axis='x', rotation=90)

# Adjust layout for clarity
fig.suptitle("Performance Comparison of Machine Learning Models for EPMA Dataset (Test Set)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
